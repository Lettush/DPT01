## **Introduction to Data Visualization**

`Data visualization` is the process of presenting information and data in a graphical format. Utilizing visual elements such as charts, graphs, and maps, data visualization tools offer an intuitive approach to interpreting trends, outliers, and patterns within data.

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

# Creating a sample DataFrame
df = pd.DataFrame({
    "a": [1,2,3,4,5],
    "b": [1,4,9,16,25]
})

# Plot the data points for (x, y)
plt.plot(df["a"], df["b"])

# Set the title of the plot
plt.title('My first plot/chart')

# Label the x-axis
plt.xlabel('A')

# Label the y-axis
plt.ylabel('B')

# Display the plot
plt.show()
```

### Types of Charts
- Bar
- Pie
- Line/Plot
- Scatter

```python
# Generating a bar chart (x, y)
plt.bar(df['a'], df['b'])

# Set the title of the plot
plt.title('My first bar chart')

# Label the x-axis
plt.xlabel('A')

# Label the y-axis
plt.ylabel('B')

# Display the plot
plt.show()
```

```python
# Generating a pie chart (x)
plt.pie(df["b"], labels=df["b"], autopct="%1.1f%%")

# Set the title of the plot
plt.title('My first pie chart')

plt.legend(df["b"])

plt.show()
```

```python
# Scatter chart (x, y)
plt.scatter(df["a"], df["b"])

# Set the title of the plot
plt.title('My first scatter chart')

# Label the x-axis
plt.xlabel('A')

# Label the y-axis
plt.ylabel('B')

plt.show()
```

```python
# Subplots (show multiple plots at once)
# 1 row and 2 columns (1, 2)
f, (ax1, ax2) = plt.subplots(1, 2)

# Line Chart
ax1.title.set_text('Line Chart')
ax1.plot(df["a"], df["b"])

# Pie Chart
ax2.title.set_text('Pie Chart')
ax2.pie(df["a"], labels=df["a"])

plt.show()
```
